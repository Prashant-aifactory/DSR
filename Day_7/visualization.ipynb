```python
# SQL - Create Table

import sqlite3

connection = sqlite3.connect("student.db")

cursor = connection.cursor()

cursor.execute("""
CREATE TABLE students(
id INTEGER,
name TEXT,
marks INTEGER
)
""")

connection.commit()

connection.close()
```

```python
#  INSERT
import sqlite3

connection = sqlite3.connect("student.db")

cursor = connection.cursor()

students = [
    (2,'Rahul',88),
    (3,'Sneha',91),
    (4,'Anjali',85),
    (5,'Kiran',78)
]

cursor.executemany(
"""
INSERT INTO students
VALUES(?,?,?)
""",
students
)

connection.commit()

connection.close()

print("Multiple Records Inserted")
```

```python
#SELECT
import sqlite3

connection = sqlite3.connect("student.db")

cursor = connection.cursor()

cursor.execute("SELECT * FROM students")

rows = cursor.fetchall()

for row in rows:
    print(row)

connection.close()
```

```python
#where
import sqlite3

connection = sqlite3.connect("student.db")

cursor = connection.cursor()

cursor.execute("""
SELECT *
FROM students
WHERE marks > 85
""")

rows = cursor.fetchall()

for row in rows:
    print(row)

connection.close()
```

```python
#update
import sqlite3

connection = sqlite3.connect("student.db")

cursor = connection.cursor()

cursor.execute("""
UPDATE students
SET marks = 90
WHERE id = 2
""")

connection.commit()

connection.close()

print("Record Updated")
```

```python
#delete
import sqlite3

connection = sqlite3.connect("student.db")

cursor = connection.cursor()

cursor.execute("""
DELETE FROM students
WHERE id = 5
""")

connection.commit()

connection.close()

print("Record Deleted")
```

```python
#order by (Ascending)
import sqlite3

connection = sqlite3.connect("student.db")

cursor = connection.cursor()

cursor.execute("""
SELECT *
FROM students
ORDER BY marks
""")

for row in cursor.fetchall():
    print(row)

connection.close()
```
