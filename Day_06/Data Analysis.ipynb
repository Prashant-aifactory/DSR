```python
# Data Analysis - NumPy Arrays
# Create a NumPy Array

import numpy as np

arr = np.array([10, 20, 30, 40, 50])

print(arr)
```

```python
# Data Analysis - NumPy Arrays
# Find the Sum

import numpy as np

arr = np.array([10, 20, 30, 40, 50])

print("Sum =", np.sum(arr))
```

```python
# Data Analysis - NumPy Arrays
# Maximum and Minimum

import numpy as np

arr = np.array([10, 20, 30, 40, 50])

print("Maximum =", np.max(arr))
print("Minimum =", np.min(arr))
```

```python
# Data Analysis - Broadcasting
# Add 10 to Every Element

import numpy as np

arr = np.array([1, 2, 3, 4, 5])

result = arr + 10

print(result)
```

```python
# Data Analysis - Pandas DataFrame
# Create a DataFrame

import pandas as pd

data = {
    "Name": ["Ram", "Shyam", "Ravi"],
    "Age": [20, 21, 22]
}

df = pd.DataFrame(data)

print(df)
```

```python
# Create a CSV File

import pandas as pd

data = {
    "Name": ["Rahul", "Priya", "Ravi"],
    "Age": [20, 21, 22],
    "Marks": [85, 90, 75]
}

df = pd.DataFrame(data)

df.to_csv("students.csv", index=False)

print("students.csv file created successfully!")
```

```python
# Data Analysis - Reading CSV
# Problem 1: Read CSV File

import pandas as pd

df = pd.read_csv("students.csv")

print(df)
```

```python
# Data Analysis - Data Preprocessing
# Remove Duplicates

import pandas as pd

data = {
    "Name": ["Ram", "Ram", "Ravi"],
    "Age": [20, 20, 22]
}

df = pd.DataFrame(data)

df = df.drop_duplicates()

print(df)
```

```python
# Data Analysis - Feature Engineering
#  Create Total Marks

import pandas as pd

data = {
    "Math": [80, 90],
    "Science": [70, 85]
}

df = pd.DataFrame(data)

df["Total"] = df["Math"] + df["Science"]

print(df)
```
