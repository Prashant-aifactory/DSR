```python
# Data Analysis - NumPy Arrays
# Create a NumPy Array

import numpy as np

arr = np.array([10, 20, 30, 40, 50])

print(arr)
```

```python
# Data Analysis - NumPy Arrays
# Problem 2: Find the Sum

import numpy as np

arr = np.array([10, 20, 30, 40, 50])

print("Sum =", np.sum(arr))
```

```python
# Data Analysis - NumPy Arrays
# Maximum and Minimum

import numpy as np

arr = np.array([10, 20, 30, 40, 50])

print("Maximum =", np.max(arr))
print("Minimum =", np.min(arr))
```

```python
# Data Analysis - Broadcasting
# Add 10 to Every Element

import numpy as np

arr = np.array([1, 2, 3, 4, 5])

result = arr + 10

print(result)
```
